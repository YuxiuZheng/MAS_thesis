# Convert JSON Files to CSV Files

```python
covid_list = []
with open('/Users/zhengyuxiu/Desktop/MAS_thesis/yelp_academic_dataset_covid_features.json') as json_file: 
    for json_object in json_file:
        covid_dict = json.loads(json_object)
        covid_list.append(covid_dict)
    
data_file = open('/Users/zhengyuxiu/Desktop/MAS_thesis/yelp_academic_dataset_covid_features.csv', 'w', newline = '')
csv_writer = csv.writer(data_file)

count = 0
for data in covid_list:
    if count == 0:
        header = data.keys()
        csv_writer.writerow(header)
        count += 1
    csv_writer.writerow(data.values())
data_file.close()
```
